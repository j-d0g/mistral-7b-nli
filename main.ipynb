```python
import pandas as pd

df = pd.read_csv('data/NLI_trial.csv')
df
```

```python
df_5 = df.head(5)
df_5
```

```python
import os

from dotenv import load_dotenv
from llm.mistral import Mistral

load_dotenv()

def generate_mistral_completion(premise, hypothesis):
    mistral = Mistral(os.getenv('MISTRAL_API_KEY'))
    system = mistral.prompt_template("system", "Your task is to solve NLI tasks. You are given a premise and a hypothesis. You need to determine whether the hypothesis is true (1) or false (0) based on the premise. You can use any information you have to make your decision, which can be based off logical reasoning or common-sense. You should return your response as a json object with the key 'thoughts' to contain your logical reasoning and deduction steps, which should come before the 'label' key, containing '1' or '0'.")
    prompt = mistral.prompt_template("user", f"Premise: {premise} Hypothesis: {hypothesis} Label:")
    mistral.add_messages([system, prompt])
    return mistral.generate_text(temperature=0.7, top_p=1.0)
    

df_5['completion'] = df_5.apply(lambda x: generate_mistral_completion(x['premise'], x['hypothesis']), axis=1)

df_5
```

```python
df['completion'] = df.apply(lambda x: generate_mistral_completion(x['premise'], x['hypothesis']), axis=1)
```

```python
df
```

```python
import json

def parse_json(json_str):
    return json.loads(json_str)

df['prediction'] = df['completion'].apply(parse_json)
df
```

```python
from llm.llama import Llama


def generate_llama_completion(premise, hypothesis):
    llama = Llama(os.getenv('REPLICATE_API_KEY'))
    system = llama.prompt_template("system", "Your task is to solve NLI tasks. You are given a premise and a hypothesis. You need to determine whether the hypothesis is true (1) or false (0) based on the premise. You can use any information you have to make your decision, which can be based off logical reasoning or common-sense. You should return your response as a json object with the key 'label' and the value '0' or '1'.")
    prompt = llama.prompt_template("user", f"Premise: {premise} Hypothesis: {hypothesis} Label:")
    llama.add_messages([system, prompt])
    return llama.generate_text(temperature=0.7, top_p=1.0)

df['completion'] = df.apply(lambda x: generate_llama_completion(x['premise'], x['hypothesis']), axis=1)
df
```
